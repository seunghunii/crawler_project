```python
import pandas as pd
from glob import glob
```

```python
dirr = 'C:/Users/shic/Desktop/shinhan_simon/google_drive/crawler_data/estate_trend/new/'
```

```python
list_files = glob(dirr + '*.csv')
list_files
```

```python
list_files
```

```python
df_list = []
cate_list = ['부동산시장','주택매매시장','주택시장_소비심리지수','주택전세시장_소비심리지수','토지시장_소비심리지수']
for file_num in range(len(list_files)):
    tmp_df = pd.read_csv(list_files[file_num],encoding='CP949')
    tmp_df['category'] = cate_list[file_num]
    tmp_df = pd.melt(tmp_df,id_vars=['시도별(1)','시도별(2)','category'])
    df_list.append(tmp_df)
```

```python
df = pd.concat(df_list,axis=0)
df
```

```python
df_pivot = df.pivot(index=['시도별(1)','시도별(2)','variable'],columns='category',values='value').reset_index()
df_pivot['variable'] = df_pivot['variable'].str.replace('. ','')
```

```python
df_pivot.columns = ['Sidonm','Signgu_Nm','Base_Ym','estate_trend','house_trade_trend','house_trend','house_rent_trend','land_trend']
df_pivot
```

```python
df_pivot.to_csv(dirr + 'df_bind.csv',index=False,encoding='UTF-8')
```
