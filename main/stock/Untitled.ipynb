```python
import pandas as pd
```

```python
jeongjae_dir = 'C:/Users/shic/Downloads/RAA_KRX업종분류현황.csv'
seunghun_dir = 'C:/Users/shic/Desktop/shinhan_simon/google_drive/crawler_data/stock/stock_preprocessed_top10.csv'
```

```python
jeongjae_df = pd.read_csv(jeongjae_dir,dtype=str)
seunghun_df = pd.read_csv(seunghun_dir,dtype=str)
```

```python
code_uni = seunghun_df['STK_STBD_CODE'].unique()
seunghun_df['BASE_YMD'] = seunghun_df['BASE_YMD'].str.replace('-','')
seunghun_df = seunghun_df[seunghun_df['BASE_YMD']!='20210730']
```

```python
jeongjae_df['종목코드'] = jeongjae_df['종목코드'].str.zfill(6)
jeongjae_df_new = jeongjae_df[jeongjae_df['종목코드'].isin(code_uni)]
jeongjae_df_new
```

```python
df_join = pd.merge(seunghun_df,jeongjae_df_new,how='left',
                   left_on=['BASE_YMD','STK_STBD_CODE'],
                   right_on=['기준일자','종목코드'])
df_join.head(5)
```

```python
df_join = df_join.dropna()
df_join.isna().sum()
```

```python
df_join_final = df_join.drop(['index','CLOSV','기준일자','MRKT_TP_1_CODE','종목코드','종목명','KRX_INDC_CODE'],axis=1)
```

```python
df_join_final = df_join_final[['BASE_YMD','KRX지표업종구분코드','업종명','STK_STBD_CODE','STBD_NM','P1DDF_RT','CURRV','HIGHV','LOWV','종가','TRDVM',
                               '전일종가대비금액','전일종가대비등락율','시가총금액']]
df_join_final.columns = ['BASE_YMD','KRX_DRTPT_INDC_TP_CODE','KRX_INDC_CODE','STK_STBD_CODE','STBD_NM','P1DDF_RT','CURRV','HIGHV','LOWV','CLOSV',
                         'TRDVM','P1D_CLOSV_DIFF_AMT','P1D_CLOSV_DIFF_UPDOWN_RT','CURRV_TAMT']
df_join_final
```

```python
df_join_final.to_csv('C:/Users/shic/Desktop/shinhan_simon/google_drive/crawler_data/stock/stock_join_final.csv',
                     index=False,encoding='UTF-8-sig')
```
