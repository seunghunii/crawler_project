```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
dirr = 'C:/Users/shic/Desktop/shinhan_simon/google_drive/crawler_data/stock/'
```

```python
df = pd.read_csv(dirr+'/stock_price_record.csv',dtype={'code':'str'})
kospi_code = pd.read_csv(dirr+'/kospi_code.csv',encoding='CP949',dtype={'종목코드':'str'})
kosdaq_code = pd.read_csv(dirr+'/kosdaq_code.csv',encoding='CP949',dtype={'종목코드':'str'})
```

```python
df = df.dropna(axis=0)
df = df.reset_index(drop=True)
df = df.rename(columns = {'code':'종목코드','type':'시장구분'})
df['날짜'] = df['날짜'].str.replace('.','',regex=False)
```

```python
df['시장구분'] = df['시장구분'].str.upper()
```

```python
code_df = pd.concat([kospi_code,kosdaq_code],axis=0)
code_df = code_df[['종목코드','종목명','시장구분','업종명']]
code_df
```

```python
df_bind = pd.merge(df,code_df,how='left',on=['종목코드','시장구분'])
```

```python
krx_code = pd.read_csv('C:/Users/shic/Desktop/tmpp_code.csv',dtype={'업종코드':'str','업종명':'str'})
```

```python
df_bind = pd.merge(df_bind,krx_code,how='left',on = '업종명')
df_bind
```

```python
df_bind['시장구분'] = df_bind['시장구분'].replace({'KOSPI':'0','KOSDAQ':'1'})
```

```python
df_bind = df_bind[['날짜','시장구분','종목코드','종목명','업종코드','종가','전일비','시가','고가','저가','거래량']]
df_bind.columns = ['BASE_YMD','MRKT_TP_1_CODE','STK_STBD_CODE','STBD_NM','KRX_INDC_CODE','CLOSV','P1DDF_RT','CURRV','HIGHV','LOWV','TRDVM']
df_bind
```

```python
df_bind['BASE_YMD'] = pd.to_datetime(df_bind['BASE_YMD'],format='%Y%m%d')
df_bind = df_bind[df_bind['BASE_YMD'] > '20111231']
```

```python
aa = df_bind.groupby(['KRX_INDC_CODE','STBD_NM'],as_index=False)['TRDVM'].mean()
aa_large = aa.groupby('KRX_INDC_CODE')['TRDVM'].nlargest(10).reset_index()
aa['index_num'] = aa.index.values
```

```python
top_stock = pd.merge(aa_large,aa,how='left',
         left_on='level_1',right_on='index_num')
top_stock
```

```python
df_bind = df_bind[df_bind['STBD_NM'].isin(top_stock['STBD_NM'])].reset_index()
```

```python
df_bind.to_csv('C:/Users/shic/Desktop/shinhan_simon/google_drive/crawler_data/stock/stock_preprocessed_top10.csv',index=0,encoding='UTF-8-sig')
```
