```python
import pandas as pd
import numpy as np
```

```python
dirr = 'C:/Users/shic/Desktop/shinhan_simon/google_drive/crawler_data/estate_registration/'
court_df = pd.read_csv('C:/Users/shic/Desktop/shinhan_simon/google_drive/crawler_data/court/RAA_법원경매통계_유승훈.csv')
trade_avg_price = pd.read_csv(dirr + 'regist_trade_avg_price(land).csv')
```

```python
regions = court_df[['SIDONM','SIGNGU_NM']].drop_duplicates()
regions
```

```python
trade_avg_price
```

```python
# 등기 데이터의 시군구에 세종특별자치시가 @로 되어 있는것을 전처리
trade_avg_price = trade_avg_price.replace({'admin_regn2_name':{'@':'세종특별자치시'}})
```

```python
trade_df_join = pd.merge(trade_avg_price,regions,how='left',
                         left_on = ['admin_regn1_name','admin_regn2_name'],
                         right_on = ['SIDONM','SIGNGU_NM'])
trade_df_join
```

```python
trade_df_join = trade_df_join[['search_start_date','admin_regn1_name','admin_regn2_name','tot1','tot2']]
trade_df_join
```

```python
trade_df_join.isna().sum()
```

```python
trade_df_join.columns = ['BASE_YM','SIDONM','SIGNGU_NM','avg_price','trade_count']
```

```python
trade_df_join.to_csv('C:/Users/shic/Desktop/shinhan_simon/google_drive/crawler_data/estate_registration/regist_trade_avg_price(land)_preprocessed.csv',
                     index=False,encoding='UTF-8-sig')
```
