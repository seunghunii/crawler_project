```python
seunghun_dirr = 'C:/Users/rsh15/Google Drive/crawler_data/court/'
shinhan_dirr = 'C:/Users/shic/Desktop/shinhan_simon/google_drive/crawler_data/court/'
df = pd.read_csv(seunghun_dirr + 'df_court_all.csv')
```

```python
for k in df.columns:
    print(df[k].unique())
```

```python
df.loc[df['물건용도_대분류']=='검색결과가 없습니다.','물건용도_대분류'] = '대분류_없음'
df.loc[df['물건용도_소분류']=='검색결과가 없습니다.','물건용도_소분류'] = '소분류_없음'
rep_cols = ['경매건수','매각건수','감정가','매각가','매각율','매각가율']
df[rep_cols] = df[rep_cols].replace('검색결과가 없습니다.','0')
```

```python
for rate_col in ['매각율','매각가율']:
    df[rate_col] = df[rate_col].str.replace(r'%','').astype(float)
    df[rate_col] /= 100
```

```python
date_bind = []
for roww in df.iterrows():
    year  = str(roww[1]['시기_년'])
    month = str(roww[1]['시기_월'])
    
    if len(month) == 1:
        date_str = year + '0' + month
    else:
        date_str = year + month
    
    date_bind.append(date_str)
```

```python
df = df[-df['SIGNGU_NM'].isin(['마산시','마산시 합포구','마산시 회원구','법인구'])]
df[df['SIGNGU_NM']=='평택군']['SIGNGU_NM'] = '평택시'
```

```python
df['date_bind'] = date_bind
```

```python
df = df[['물건용도_대분류','물건용도_소분류','시/도','시/군/구','date_bind','경매건수','매각건수','감정가','매각가','매각율','매각가율']]
df.columns = ['ACTN_THNG_PRPS_BCLAS_NM','ACTN_THNG_PRPS_SCLAS_NM','SIDONM','SIGNGU_NM','BASE_YM','ACTN_ITMN','DISP_ITMN','ESTM_PRICE',
              'DISP_PRICE','DISP_RT','DISP_VLUE_RT']
```
