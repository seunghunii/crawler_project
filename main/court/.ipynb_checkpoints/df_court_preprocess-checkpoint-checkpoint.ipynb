```python
import pandas as pd
import numpy as np
import re
import warnings
warnings.filterwarnings('ignore')
```

```python
seunghun_dirr = 'C:/Users/rsh15/Google Drive/crawler_data/court/'
shinhan_dirr = 'C:/Users/shic/Desktop/shinhan_simon/google_drive/crawler_data/court/'
df = pd.read_csv(shinhan_dirr + 'df_court_bind.csv')
```

```python
for k in df.columns:
    print(df[k].unique())
```

```python
# '검색결과가 없습니다.' 결과의 지역, 시기 데이터와 물건용도 분류를 조합하여 모두 0인 데이터를 만들어 삽입.
emp_list = []
for roww in df.loc[df['물건용도_대분류']=='검색결과가 없습니다.',['시/도','시/군/구','시기_년','시기_월']].iterrows():
    
    tmp_df = df.loc[df['물건용도_대분류']!='검색결과가 없습니다.',['물건용도_대분류','물건용도_소분류']].drop_duplicates()
    tmp_df[['경매건수','매각건수','감정가','매각가','매각율','매각가율']] = 0
    tmp_df['시/도'] = roww[1]['시/도']
    tmp_df['시/군/구'] = roww[1]['시/군/구']
    tmp_df['시기_년'] = roww[1]['시기_년']
    tmp_df['시기_월'] = roww[1]['시기_월']
    emp_list.append(tmp_df)

emp_df = pd.concat(emp_list,axis=0)
```

```python
df = pd.concat([df,emp_df],axis=0).sort_values(['시/도','시/군/구','시기_년','시기_월'])
df = df[df['물건용도_대분류']!='검색결과가 없습니다.']
```

```python
# 인덱스를 통해 저장했기 때문에 년,월 데이터를 변환.
df['시기_월'] += 1 # 0부터 시작하는 인덱스에 1씩 더해서 월 데이터로 변환.
df['시기_년'] = df['시기_년'].replace(dict(zip(
    range(0,10,1),
    range(2021,2011,-1))))
```

```python
for rate_col in ['매각율','매각가율']:
    df[rate_col] = df[rate_col].str.replace(r'%','').astype(float)
    df[rate_col] /= 100
    df[rate_col] = np.round(df[rate_col],3)
```

```python
date_bind = []
for roww in df.iterrows():
    year  = str(roww[1]['시기_년'])
    month = str(roww[1]['시기_월'])
    
    if len(month) == 1:
        date_str = year + '0' + month
    else:
        date_str = year + month
    
    date_bind.append(date_str)

df['date_bind'] = date_bind
```

```python
# 시/도, 시/군/구명 정리
df = df[-df['시/군/구'].isin(['마산시','마산시 합포구','마산시 회원구','법인구'])]
df['시/군/구'] = df['시/군/구'].replace({'평택군':'평택시',
                                       '청원군':'청주시',
                                       '진해시':'창원시 진해구',
                                       '온양시':'아산시',
                                       '연기군':'세종특별자치시',
                                       '없음':'세종특별자치시',
                                       '여주군':'여주시',
                                       '안성군':'안성시',
                                       '당진군':'당진시',
                                       '부천시 남구':'부천시',
                                       '부천시 소사구':'부천시',
                                       '부천시 오정구':'부천시',
                                       '부천시 원미구':'부천시',
                                       '부천시 중구':'부천시'})
```

```python
df = df.reset_index(drop=True)
```

```python
for col in ['경매건수','매각건수','감정가','매각가']:
    df[col] = df[col].astype(float)
```

```python
df = df.groupby(['물건용도_대분류','물건용도_소분류','시/도','시/군/구','date_bind'],as_index=False).agg({'경매건수':'sum','매각건수':'sum','감정가':'sum',
                                                                                                    '매각가':'sum','매각율':'mean','매각가율':'mean'})
```

```python
df = df[['물건용도_대분류','물건용도_소분류','시/도','시/군/구','date_bind','경매건수','매각건수','감정가','매각가','매각율','매각가율']]
df.columns = ['ACTN_THNG_PRPS_BCLAS_CODE','ACTN_THNG_PRPS_SCLAS_CODE','SIDONM','SIGNGU_NM','BASE_YM','ACTN_ITMN','DISP_ITMN','ESTM_PRICE',
              'DISP_PRICE','DISP_RT','DISP_VLUE_RT']
```

```python
df = df.replace({
    'ACTN_THNG_PRPS_SCLAS_CODE':{'아파트':'소계','연립주택,다세대':'소계','자동차,중기':'소계',
                                 '기타':'소계','전체':'소계'}})
```

```python
df[df['ACTN_THNG_PRPS_BCLAS_CODE'].isin(['아파트','연립주택,다세대','자동차,중기','기타','전체'])]['ACTN_THNG_PRPS_SCLAS_CODE'] = '소계'
```

```python
df['ACTN_THNG_PRPS_BCLAS_CODE'] = df['ACTN_THNG_PRPS_BCLAS_CODE'].replace({'전체':'0','아파트':'1','단독주택,다가구주택':'2',
                                                                       '연립주택,다세대':'3','대지,임야,전답':'4','자동차,중기':'5',
                                                                       '상가,오피스텔,근린시설':'6','기타':'7'})
```

```python
np.unique(df['ACTN_THNG_PRPS_BCLAS_CODE'],return_counts=True)
```

```python
df['ACTN_THNG_PRPS_SCLAS_CODE'] = df['ACTN_THNG_PRPS_SCLAS_CODE'].replace({'소계':'0','겸용':'1','단독주택':'2','다가구주택':'3','대지':'4',
                                                                       '임야':'5','전답':'6','상가':'7','오피스텔':'8','근린시설':'9'})
```

```python
np.unique(df['ACTN_THNG_PRPS_SCLAS_CODE'],return_counts=True)
```
