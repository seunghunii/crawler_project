```python
import pandas as pd
from bs4 import BeautifulSoup
import urllib.request as reqests
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

## Task1: st johns hotel

```python
urll = 'https://www.tripadvisor.co.kr/Hotel_Review-g317126-d13397307-Reviews-or{}-St_John_s_Hotel-Gangneung_Gangwon_do.html'
```

```python
reviewers = []
title = []
review = []
date = []
data_list = []
urll = 'https://www.tripadvisor.co.kr/Hotel_Review-g317126-d13397307-Reviews-or{}-St_John_s_Hotel-Gangneung_Gangwon_do.html'
for k in range(5,30,5):
    k_url = urll.format(k)
    
    html = reqests.urlopen(k_url)
    doc = BeautifulSoup(html,'html.parser')
    
    end1 = len(doc.find_all('a',class_='ui_header_link _1r_My98y'))
    reviewers.extend([doc.find_all('a',class_='ui_header_link _1r_My98y')[n].text for n in range(0,end1)])
    
    end2 = len(doc.find_all(class_='ocfR3SKN'))
    title.extend([doc.find_all(class_='ocfR3SKN')[n].text for n in range(0,end2)])
    
    end3 = len(doc.find_all(class_='IRsGHoPm'))
    review.extend([doc.find_all(class_='IRsGHoPm')[n].text for n in range(0,end3)])
    
    for t in doc.find_all(class_='_355y0nZn'):
        t.decompose()
    
    end4 = len(doc.find_all(class_='_34Xs-BQm'))
    date.extend([doc.find_all(class_='_34Xs-BQm')[n].text for n in range(0,end4)])
    
    tmp_df = pd.DataFrame({
        'reviewers':reviewers,
        'title':title,
        'review':review,
        'date':date
    })
    data_list.append(tmp_df)
```

```python
pd.concat(data_list,axis=0)
```

```python
driver = webdriver.Chrome(dirr)
driver.get(url = 'https://www.opentable.com/new-york-city-restaurants')
driver.maximize_window()
    
driver.find_element_by_css_selector('#downshift-0-input').send_keys('korean')
driver.find_element_by_class_name('_3E4Lqktx3MqqARumeR3NnT').click()
    
click = driver.find_element_by_class_name('_3E4Lqktx3MqqARumeR3NnT')
click.click()
    
for _ in range(0,10):
    driver.find_element_by_css_selector('body').send_keys(Keys.PAGE_DOWN)
        
restaurants = driver.find_elements_by_class_name('_1e9PcCDb012hY4BcGfraQB')

names = []
reviews = []
ratings = []
tag1 = []
tag2 = []
tag3 = []

i = 1
for r in restaurants:
    main_window = driver.current_window_handle
    r.click()
    
    driver.switch_to_window(driver.window_handles[1])
    
    sleep(3)
    
    doc = BeautifulSoup(driver.page_source,'html.parser')
    driver.close()
    
    tag = doc.find_all(id='toptags')
    name = doc.find(class_='a6481dc2 _4a4e7a6a').text
    names.extend([doc.find(class_='a6481dc2 _4a4e7a6a').text])
    
    rev = doc.find(id='total-reviews')
    rat = doc.find(id='star-rating')
    
    try:
        reviews.extend([rev.find(class_='df8add00').text])
    except:
        reviews.extend(['No Reviews'])
        
    try:
        ratings.extend([rat.find(class_='df8add00').text])
    except:
        ratings.extend(['No Ratings'])
        
    try:
        tag1.extend(tag[0].text)
    except:
        tag1.extend(['No Tags'])
    
    try:
        tag2.extend(tag[1].text)
    except:
        tag2.extend(['No Tags'])
        
    try:
        tag3.extend(tag[2].text)
    except:
        tag3.extend(['No Tags'])
    
    print(str(i) + '. ' + name)
    i += 1
    
    driver.switch_to_window(main_window)
    sleep(3)
    
    print('END')
```
