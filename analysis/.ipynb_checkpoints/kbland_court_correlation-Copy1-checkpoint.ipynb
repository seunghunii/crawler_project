```python
court = court[court['ACTN_THNG_PRPS_BCLAS_CODE']==1]
court_kangwon = court.groupby(['SIDONM','BASE_YM'],as_index=False).agg({
    'ACTN_ITMN':'sum','DISP_ITMN':'sum',
    'ESTM_PRICE':'sum','DISP_PRICE':'sum',
    'DISP_RT':'mean','DISP_VLUE_RT':'mean'
})
court_kangwon
```

```python
court_kangwon['BASE_YM'] = pd.to_datetime(court_kangwon['BASE_YM'],format='%Y%m')
court_kangwon = court_kangwon[court_kangwon['BASE_YM']<'2021-07-01']
```

```python
kb_cols = kb_kangwon.columns.values.tolist()
court_cols = court_kangwon.columns.values.tolist()
```

```python
print(kb_cols)
print(court_cols)
```

```python
np.corrcoef([kb_kangwon[kb_cols[4]].values,
             court_kangwon[court_cols[6]].values])
```

```python
corr_result = []
for kbb in kb_cols[1:]:
    for courtt in court_cols[1:]:
        x = kb_kangwon[kbb].values
        y = court_kangwon[courtt].values
        
        np_coef = np.corrcoef([x,y])
        sc_pear = pearsonr(x,y)
        
        corr_result.append(pd.DataFrame({
            'kb_cols' : kbb,
            'court_cols' : courtt,
            'np_corr' : np_coef[0,1],
            'sc_corr' : sc_pear[0],
            'sc_pval' : round(sc_pear[1],3)
        },index=[0]))
```

```python
corr_df = pd.concat(corr_result)
corr_df[corr_df['sc_pval']<0.05]
```
